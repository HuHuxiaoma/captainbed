```python
import numpy as np # 加载numpy工具库并给它取个别名为np，后面就可以通过np来调用numpy工具库里面的函数了。numpy是python的一个科学计算工具库，
                    # 除了前面文章中提到的它可以用来进行向量化之外，它还有很多常用的功能。非常非常常用的一个工具库！
import matplotlib.pyplot as plt # 这个库是用来画图的

import h5py # 这个库是用来加载训练数据集的。我们数据集的保存格式是HDF。Hierarchical Data Format(HDF)是一种针对大量数据进行组织和存储的
            #  文件格式,大数据行业和人工智能行业都用它来保存数据。
import skimage.transform as tf # 这里我们用它来缩放图片
# 加载我们自定义的工具函数
from testCases import *
from dnn_utils import *
#这是jupyter notebook里的命令, 意思是将那些用matplotlib绘制的图显示在页面里而不是弹出一个窗口
%matplotlib inline 

def load_data():
    train_data_set = h5py.File("datasets/train_catvnoncat.h5","r") #加载需要训练的数据
    train_set_x_orig = np.array(train_data_set["train_set_x"][:])#加载训练数据x对应的样本特征数据
    train_set_y_orig = np.array(train_data_set["train_set_y"][:]) #载训练数据对应样本的y 标签数据
    
    test_data_set = h5py.File("datasets/test_catvnoncat.h5","r")#加载测试的数据
    test_set_x_orig = np.array(test_data_set["test_set_x"][:])#加载测试数据x对应的特征数据
    test_set_y_orig = np.array(test_data_set["test_set_y"][:])#加载测试数据y对应的标签数据
    
    
    classnames = np.array(test_data_set["list_classes"][:])#加载标签类别数据，这里的类别只有两种，1代表有猫，0代表无猫
    

    train_set_y_orig = train_set_y_orig.reshape(1 ,train_set_y_orig.shape[0])#重新组织数组的维度 从（209,）变为 （1，209）1*209的维度
    test_set_y_orig =  test_set_y_orig.reshape(1 ,test_set_y_orig.shape[0])#重新组织数组的维度 从（50,）变为 （1，50）1*50的维度 方便计算
    
#     print (train_set_x_orig.shape)
#     print (train_set_y_orig.shape)
#     print (test_set_x_orig.shape)
#     print (test_set_y_orig.shape)
    
    return train_set_x_orig,train_set_y_orig,test_set_x_orig,test_set_y_orig,classnames
# index=30
# plt.imshow(train_set_x_orig[index])    
# print ("这张图片的标签为" + str (train_set_y_orig[:, index]) + ", 这是一个'" + classnames[np.squeeze(train_set_y_orig[:, index])].decode("utf-8") +  "' 图片.")
train_x_orig, train_y, test_x_orig, test_y, classes = load_data()

m_train = train_x_orig.shape[0] # 训练样本的数量
m_test = test_x_orig.shape[0] # 测试样本的数量
num_px = test_x_orig.shape[1] # 每张图片的宽/高

# 为了方便后面进行矩阵运算，我们需要将样本数据进行扁平化和转置
# 处理后的数组各维度的含义是（图片数据，样本数）
train_x_flatten = train_x_orig.reshape(train_x_orig.shape[0], -1).T
test_x_flatten = test_x_orig.reshape(test_x_orig.shape[0], -1).T 

# 下面我们对特征数据进行了简单的标准化处理（除以255，使所有值都在[0，1]范围内）
train_x = train_x_flatten/255.
test_x = test_x_flatten/255.
def init_parameters(dims):
    #dims -- 这个list列表里面，包含了每层的神经元个数。
    #例如，dims=[5,4,3]，表示第一层有5个神经元，第二层有4个，最后一层有3个神经元
    
    np.random.seed(1)
    parameters = {}
    #神经层数
    L = len(dims)
    print(dims)
    for l in range (1,L):
        parameters["W"+ str(l)] = np.random.randn(dims[l],dims[l-1])/np.sqrt(dims[l-1])
        parameters["b"+ str(l)] = np.zeros((dims[l],1))
   
    return parameters

def front_step_one(A,W,b):#A同X 上一层的A 作为X输入
    
    #d单个样本训练的前向传播
    Z = np.dot(W, A) + b
    
    cache = (A,W,b)#缓存起来以便反向传播时需要
    return Z,cache

def front_step_two(A,W,b,active):
    Z, linear_cache = front_step_one(A, W, b)
    if active == "relu":
        A = relu(Z)
    elif  active =="sigmoid":
        A = sigmoid(Z)
    cache = (linear_cache ,Z)#缓存起来以便反向传播时需要
    return A ,cache 
def front(X,parameters):
    
    # 获取参数列表的长度，这个长度的一半就是神经网络的层数。
    # 为什么是一半呢？因为列表是这样的[w1,b1,w2,b2...wl,bl],里面的w1和b1代表了一层
    L = len(parameters) // 2 #
    caches = []
    A = X #第一次层的A 就是X 之后迭代A 传入下一层
    for l in range(1,L):#range(1,3 )表示 1 - 2 的整数 不包括 3
        # 前面的激活函数用relu 所以不包括最后一层
        A ,cache = front_step_two(A,parameters["W" + str(l)],parameters["b" + str(l)],active='relu')
        caches.append(cache)
    #单独计算最后一层
    AL , cache = front_step_two(A,parameters["W" + str(L)],parameters["b" + str(L)],active='sigmoid')    
    caches.append(cache) # 之所以缓存cache参数 是因为反向传播的时候 需要用到 当前的A
    
    return AL ,caches

def compute_cost(AL, Y):
       
    m = Y.shape[1]
    cost = (-1 / m) * np.sum(np.multiply(Y, np.log(AL)) + np.multiply(1 - Y, np.log(1 - AL)))
    
    cost = np.squeeze(cost)# 确保cost是一个数值而不是一个数组的形式
    assert(cost.shape == ())
    
    return cost

def back_line(dZ,cache):
    
    #cache (A ,W,b)
    m = cache[0].shape[1] #A的样本数
    dW = np.dot(dZ,cache[0].T)/ m
    db = np.sum(dZ) / m
    
    dA = np.dot(cache[1].T,dZ)# 前一个dA 
    return dA,dW,db
    
def back_dz(dA,cache,active):
    liner_cache,active_cache = cache #cache(liner_cache ,Z)
    if active == "relu":
        dZ = relu_backward(dA,active_cache)
    elif active == "sigmoid":
        dZ = sigmoid_backward(dA,active_cache)
        
    dA,dW,db = back_line(dZ,liner_cache)
    return dA,dW,db
    
def back(AL,Y,caches):
    
    L = len(caches)
#     print(caches)
    Y = Y.reshape(AL.shape)
    grads = {}
    #最后一层
#     dAL = - (np.divide(Y, AL) - np.divide(1 - Y, 1 - AL)) //之前这里公式写错 找了好久的问题
    dAL = -(np.divide(Y,AL) - np.divide(1-Y,1-AL))
    curr_cache = caches[-1]
    grads["dA" + str(L - 1)],grads["dW" + str(L)],grads["db" + str(L)] = back_dz(dAL,curr_cache,active='sigmoid')
     
    for c in reversed(range(1, L)): #倒叙循环 L-1 L-2,L-3 ...1
          grads["dA" + str(c - 1)],grads["dW" + str(c)],grads["db" + str(c)] = back_dz(grads["dA" + str(c)],caches[c-1],active='relu')
    

    return grads
    
def update_parameters(parameters,grads,learning_rate=0.0075):
    
    L = len(parameters) //2
    for l in range(1,L+1): #这里+1 主要是因为 之前初始化的时候按照len(dims) 开始 实际len(parameter)已经少了一个，所有要+ 1回来
        parameters["W" + str(l)] = parameters["W" + str(l)] - learning_rate * grads["dW" + str(l)]
        parameters["b" + str(l)] = parameters["b" + str(l)] - learning_rate * grads["db" + str(l)]
        
    return parameters
def model(X,Y,dims,t_number,learning_rate,print_cost=False):
    
    costs = []
    parameters = init_parameters(dims)
    for i in range(0,t_number):
        
        #前向传播
        
        AL,caches = front(X,parameters)
#         print(AL)
        #陈本
        cost = compute_cost(AL,Y)
        #反向传播、
        grads = back(AL,Y,caches)
        #更新参数   这里迭代参数 
        parameters = update_parameters(parameters,grads,learning_rate)
        # 打印出成本
        if i % 100 == 0:
            if print_cost and i > 0:
#                 print(cost)
                print ("训练%i次后成本是: %f" % (i, cost))
            costs.append(cost)
        
    plt.plot(np.squeeze(costs))
    plt.ylabel('cost')
    plt.xlabel('iterations (per tens)')
    plt.title("Learning rate =" + str(learning_rate))
    plt.show()
    return parameters
dims = [12288, 20, 7, 5, 1]
# parameters= model(train_x,train_y,dims,t_number=2000,learning_rate=0.0075,print_cost=True)
# print(parameters)

def predict(X,parameters):   
    m = X.shape[1]
    p = np.zeros((1,m))
    
    # 进行一次前向传播，得到预测结果
    A, caches = front(X, parameters)
   
    # 将预测结果转化成0和1的形式，即大于0.5的就是1，否则就是0
    for i in range(0, A.shape[1]):
        if A[0,i] > 0.5:
            p[0,i] = 1
        else:
            p[0,i] = 0
        
    return p
# print(parameters)
pred_train = predict(train_x,parameters)
print("预测准确率是: "  + str(np.sum((pred_train == train_y) / train_x.shape[1])))
pred_test = predict(test_x,parameters)
print("预测准确率是: "  + str(np.sum((pred_test == test_y) / test_x.shape[1])))
my_image="images/hd.jpg"
image = np.array(plt.imread(my_image));
# print(image)
image_shape = tf.resize(image,(num_px,num_px),mode='reflect').reshape((1,num_px*num_px*3)).T
Y_image = predict (image_shape,parameters)
plt.imshow(image)
print("预测结果：" + str(int(np.squeeze(Y_image))))
```
